*After analysing the data with Pixar, you can use this script to customize and replot some of the maps. Refer user manual or video tutorial for more details.

#### To run the script:
1. Copy your PixAR 'Results_<your filename>' folder to the PixAR directory.
2. Change plot features in 'Plot Customization' and 'Clustermap Customization' cells.
3. Select **Kernel > Restart & Run all**
4. You will be prompted to enter the Results folder name.
5. Replotted data will be stored in 'Replotted Data' folder inside your 'Results_<your filename>' folder

## Plot Customization

**List of colors for background:** https://matplotlib.org/stable/gallery/color/named_colors.html

```python
#change time (in h) for phase map plotting if required. If set to auto, time from your pixar analysis will be used. 
phasemap_time = 'auto' #set to 'auto' or any time value (without quotes).

#set colors
periodmap_color = "jet" #color scale for period maps
phasemap_color = "hsv" #color scale for phase maps (preferably use cyclic colormaps)
map_background_color = 'black'#background color for all plots mentioned above

#set min and max periods for period maps.If set to auto, time from your pixar analysis will be used.
#Note: this does not change period range for analysis or other results. It just changes period range considered for the plots.
maps_minper = 'auto' #set to 'auto' or any time value (without quotes).
maps_maxper = 'auto' #set to 'auto' or any time value (without quotes).

#set min and max amplitude for each of the three maps separately
#Note: this does not change period range for analysis or other results. It just changes period range considered for map plots.
cwt_minamp = 'auto' #cwt minimum amp
cwt_maxamp = 'auto'  #cwt maximum amp

metacyc_minamp = 'auto' #metacycle minimum amp
metacyc_maxamp = 'auto' #metacycle maximum amp

xov_minamp = 'auto' #crossover minimum amp
xov_maxamp = 'auto' #crossover maximum amp

#set dpi for saving figures
dpi = 300
```
